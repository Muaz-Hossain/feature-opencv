```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/love.png')
if img is None:
  print('Could not open or find the image')
  exit

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
edges = cv2.Canny(gray, 100, 200)
cv2_imshow(edges)
ret,thrus = cv2.threshold(edges, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(thrus)
con,hrechy = cv2.findContours(thrus,cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
con_img = cv2.drawContours(img.copy(), con, -1, (0, 255, 0), 3)
print(len(con))


cv2.waitKey(0)
cv2.destroyAllWindows()
```
